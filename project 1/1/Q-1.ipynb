```python
#import library
import random #To create numbers and select digits from the allowed list, for example, for base 2, only digits 0 and 1 are allowed and must be used.
import re #To clean the number and calculate its value
```

```python
#Here we give the list of allowed numbers and say choose randomly from it and we do this action for <K> times.
def bulid_number_temp(list_of_number : list, k):
    result = "" #Define an empty string to add the selected digits to the end each time
    for repeat in range(0, k): #Do this operation for <K> times so that the length of the number is correct
        result += random.choice(list_of_number)
    
    return str(result)  #return it
```

```python
#In this function we normalize the numbers with floating point conditions
#I will explain step by step
def build_number(base, k):
    Allowed_number = range(0, base) #Store the numbers allowed to be used in a list using the (range) function.
    Allowed_numbers = [str(element) for element in Allowed_number] #Convert them to characters for use in the functions you defined

    numbers = []  #Define an empty list to store the obtained digits in the normal floating point system

    while True: #A loop is always correctly defined, and its stopping condition is written at the end, which means the number of bottles produced

        number_temp = bulid_number_temp(Allowed_numbers, k) #Use the above function to generate the number
        if (number_temp[0] == "0") or (number_temp  in numbers): #The bet is checked, if the first digit is zero or the generated number is in the list of numbers, then the ring will do one more round
          continue
        
        numbers.append(number_temp) #Otherwise, the loop continues and is stored in our list of numbers

        if len(numbers) == ((base - 1) * (base ** (k - 1))): #The condition of our stop is that the produced arahm is equal to the formula we have from the lesson
            break
        #Note: Multiplication by 2 is not taken into account because positive and negative are not important for us at the moment
    return numbers #return it
```

```python
#Creating complete numbers (decimal and exponents)
#Inputs of the function: the output of the upper / base function and (lower) and (upper)
def build_power(numbers : list, base, lower, upper):
  number_fainal = []  #Define an empty list to store the final numbers

  for number in numbers: #We do a <iterate> in the list of numbers output from the above function
    for power in range(lower, upper + 1): #For each number, the number of powers that we have created according to the inputs of the function
      number_fainal.append(f"(0/{number}) * {base}^{power}") #We make them in standard form with <f string>
  

  return number_fainal  #return it
```

```python
#This function is for separating the numerical and decimal part
def clean_num(number : str): #The input to each function is the number to be cleaned
  num , pow = number.split("^") #The <split> function divides the desired string into two parts and the basis of that division is the input of the <split> function
  number_usefull , temp = num.split("*")
  temp, number_usefull = number_usefull.split("/")
  del temp
  number_usefull = re.sub('\)', '', number_usefull)

  result = [number_usefull[:-1], pow]

  return (result) #return it
```

```python
#This function is used to calculate the amount of generated digits
#The inputs of the function are the generated list of numbers and its base
def calcute_value(numbers : list, base):
  result = []  #Define an empty list to store the value
  for element in numbers:  #One has <iterate> in her list of numbers
    temp = clean_num(element)  #Using the <clean_num> function, define the number into the main number part and its power
    #Separation of digit and power digit
    number = temp[0] 
    pow = temp[1] 

    sum = 0 #Define a variable to calculate the sum


    #In the list of number one >> and calculated its value using the >> function by means of indexes

    for index , digit in enumerate(number):
      sum += int(digit) * (base ** (-(index + 1))) 

    #The amount of my power has also multiplied
    fainal = sum * (base ** int(pow)) 
    result.append(f"  +/- {element}  <---------------------->  value : +/-  {fainal}") #Save the number along with their value
  

  return result #return it
```

```python
#for show
def show(numbers : list):
  for index, element in enumerate(numbers):
    print(f"{index + 1}) {element}")
    print("*********************")
```

```python
#To organize these functions, another function called <main> has been defined
def main(base, k, lower, upper):
  numbers = build_number(base, k) #Building numbers (without decimal form and without powers)
  numbers = build_power(numbers, base, lower, upper)#Storage with power
  numbers = calcute_value(numbers , base)#Calculate their value and save them in the final form
  show(numbers)##the show
```

```python
#see result
main(2, 5 ,-5, 5)
```
