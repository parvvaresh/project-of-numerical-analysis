```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definition of the sum of differences function
def F(X, Y):
  if len(Y) == 2:
    return (Y[1] - Y[0]) / (X[1]- X[0])
  else:
    return (F(Y[ : -1], X[ : -1]) - F(Y [1 : ], X[1 : ])) / (X[-1] - X[0])   #The basis of the function is recursive
```

```python
#Implementation of Newton's interpolation based on its formula
def newtonInterpolation(X, Y, x):
  sum = Y[0]
  for i in range(1, len(X)):
    temp = 0
    for j in range(0, i):
      temp += x - X[j]
    sum += (temp * F(X[ : i + 1], Y[ : i + 1]))
  return sum
```

```python
def calcute(X, Y, x):
  y = []
  for element in x:
    y.append(newtonInterpolation(X, Y, element))
  return y
```

```python
def main(X, Y):
  x = list(np.linspace(-5, 5, 1000))  #Generate a thousand points to draw a graph
  y = calcute(X, Y, x)
  plt.plot(x, y)
```

```python
X = [-1, 0, 1, 2]
Y = [1, -3, -2, 5]
```

```python
main(X, Y)
```
