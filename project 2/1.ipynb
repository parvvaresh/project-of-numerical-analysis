```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def founction(x):
  return 1 / (1 + (x ** 2))
```

```python
#Generate an arbitrary number of points between intervals
def make_points(start, end, number_of_points):
  h = abs(start - end) / (number_of_points - 1)
  result = []
  for i in range(0, number_of_points):
    result.append(start + (i * h))
  return result
```

```python
#Implementation of function L
def l(i, points, x, number_of_points):
  result = 1
  x_i = points[i]
  for j in range(0, number_of_points):
    if(i != j):
      result *= (x - points[j]) / (x_i - points[j])
  return result * founction(x)
```

```python
def calcute_y_main(points):
  result = []
  for element in points:
    result.append(founction(element))
  return result
```

```python
def laqrage(number_of_points : int, start = -5, end = 5):
  points = make_points(start, end, number_of_points)
  test_point = test = list(np.linspace(start, end, 1000))
  y_main = calcute_y_main(test_point)
  y_approximation = []
  for x in test_point:
    sum = 0
    for i in range(0, number_of_points):
      sum += l(i, points, x, number_of_points) 
    y_approximation.append(sum) 
  
  plt.plot(test_point, y_main, color="red")
  plt.show()
  plt.plot(test_point, y_approximation, color="blue")
  plt.show()
```

```python
laqrage(25)
```
